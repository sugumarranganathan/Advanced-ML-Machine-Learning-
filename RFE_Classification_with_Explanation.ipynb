```python
# =========================================================
# 2. TRAIN / TEST SPLIT + FEATURE SCALING
# =========================================================
def split_scalar(indep_X, dep_Y):
    """
    Split dataset into train/test and apply StandardScaler.
    """

    X_train, X_test, y_train, y_test = train_test_split(
        indep_X, dep_Y, test_size=0.25, random_state=0
    )

    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)

    return X_train, X_test, y_train, y_test


# =========================================================
# 3. COMMON PREDICTION + METRICS FUNCTION
# =========================================================
def cm_prediction(classifier, X_test, y_test):
    """
    Predict test data and return evaluation metrics.
    """

    y_pred = classifier.predict(X_test)

    cm = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)

    return classifier, accuracy, report, X_test, y_test, cm


# =========================================================
# 4. LOGISTIC REGRESSION
# =========================================================
def logistic(X_train, y_train, X_test, y_test):
    classifier = LogisticRegression(random_state=0, max_iter=1000)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)


# =========================================================
# 5. SVM LINEAR
# =========================================================
def svm_linear(X_train, y_train, X_test, y_test):
    classifier = SVC(kernel='linear', random_state=0)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)

```

```python
# =========================================================
# 6. SVM NON-LINEAR (RBF)
# =========================================================
def svm_NL(X_train, y_train, X_test, y_test):
    classifier = SVC(kernel='rbf', random_state=0)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)


# =========================================================
# 7. NAIVE BAYES
# =========================================================
def Naive(X_train, y_train, X_test, y_test):
    classifier = GaussianNB()
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)


# =========================================================
# 8. KNN
# =========================================================
def knn(X_train, y_train, X_test, y_test):
    classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)


# =========================================================
# 9. DECISION TREE
# =========================================================
def Decision(X_train, y_train, X_test, y_test):
    classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)


# =========================================================
# 10. RANDOM FOREST
# =========================================================
def random_forest(X_train, y_train, X_test, y_test):
    classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)
```

```python
def rfeFeature(indep_X, dep_Y, n):
    """
    Apply RFE using multiple models and return transformed feature sets.
    """
    rfelist = []

    # Scale data before RFE (important for Logistic Regression / SVM)
    sc = StandardScaler()
    indep_X_scaled = sc.fit_transform(indep_X)

    # Models used for RFE
    log_model = LogisticRegression(solver='lbfgs', max_iter=5000)
    rf_model = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
    dt_model = DecisionTreeClassifier(criterion='gini', max_features='sqrt', splitter='best', random_state=0)
    svc_model = SVC(kernel='linear', random_state=0)

    rfemodellist = [log_model, svc_model, rf_model, dt_model]

    for model in rfemodellist:
        print("RFE Model:", model)

        selector = RFE(estimator=model, n_features_to_select=n)
        selector.fit(indep_X_scaled, dep_Y)

        selected_features = selector.transform(indep_X_scaled)
        rfelist.append(selected_features)

    return rfelist
```

```python
def rfe_classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf):

    df_rfe = pd.DataFrame(index=['Logistic', 'SVM_Linear', 'SVM_NL', 'KNN', 'Naive', 'Decision', 'RandomForest'])
    for i, model_name in enumerate(['RFE_Log', 'RFE_SVM_Linear', 'RFE_RF', 'RFE_DT']):
        df_rfe[model_name] = [acclog[i], accsvml[i], accsvmnl[i], accknn[i], accnav[i], accdes[i], accrf[i]]
    return df_rfe
```
